# Jupyter Notebook
Jupyter is used by a lot of scientist to describe easily the result or to explain how to use some specific program that they have created.

Other documentation, with examples and other code on (https://github.com/ipython/ipython/wiki)

## Installation

The installation can be done with Anaconda but we will use pip, so

    pip install notebook

If you have different version of Python use the following

    python3.10 -m pip install notebook

and to lunch it we go to the terminal and run

    jupiter notebook

In this case it will start a OFFLINE SERVER (No need of internet connection) that has the same interface of a generic browser interface. I can navigate in folder and create new one. **Note that the compilation is not necessary Top to Bottom, but is in the order that I run the cell**. But we can use also the command RUN ALL to run all cell from Top to Bottom.

## ShortCut

Very useful are also the KEYBOARD SHORTCUT.

    CTRL + ENTER -> Run the cell

    SHIFT + ENTER -> Run the cell and insert below

    ALT + ENTER -> Run the cell and insert below also if there is cell below

    TAB -> To complete variable or function name

    SHIFT + TAB -> Access to info of a function

## MARKDOWN SYNTAX

In the following we report an example with most popular choice of syntax. To make a TITLE we use

    # Title

For the SECTION

    ## Section

For the SUBSECTION

    ### Subsection

To make ITALICIZED text we use (*Example of italicized*)

    *text to italicize*

To make BOLD we use (**example for bold signal**)

    **text to bold**

## Example code in Python



```python
'''
Simple demonstration of matplotlib
'''

import numpy as np
import matplotlib.pyplot as plt

N=50
# Create vector
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2

# Plot random scatter point with random color and random Area value
plt.scatter(x, y, s=area, c=colors)
plt.show()
```
